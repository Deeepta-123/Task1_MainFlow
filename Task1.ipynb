```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


dataset_path="student-mat.csv"

#DataFrame
data=pd.read_csv("student-mat.csv",sep=";")
print("Data loaded successfully")

#Data Exploration
print(data.head())
print("\nDataset Info:")
print(data.info)

#Data Cleaning
print("\nMissing Values Check :")
print(data.isnull().sum())
print(data.dtypes)
print(data.shape)

#Remove Duplicates
data=data.drop_duplicates()

#Data Analysis
#What is the average score in math (G3)?
avg_score=data['G3'].mean()
print(f"\nAverage Math Score(G3):{avg_score:.2f}")

#2. How many students scored above 15 in their final grade (G3)?
students_15=len(data[data['G3']>15])
print(f"\nNumber of students scoring above 15:{students_15}")

#3. Is there a correlation between study time (study time) and the final grade (G3)?
correlation=data['studytime'].corr(data['G3'])
print(f"Correlation between study Time and Final Grade:{correlation:.2f}")

#4. Which gender has a higher average final grade (G3)?
avg_grade=data.groupby('sex')['G3'].mean()
print("\nAverage Final Grade by Gender")
print(avg_grade)

#Data Visualization
# Plot a histogram of final grades (G3).
plt.figure(figsize=(8,5))
plt.hist(data['G3'],bins=10,color='skyblue', edgecolor='black')
plt.title("Distribution of final Grade(G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()

#2. Create a scatter plot between study time (study time) and final grade (G3).
plt.figure(figsize=(8,5))
sns.scatterplot(data=data,x='studytime',y='G3',hue='sex')
plt.title("StudyTime vs FinalGrade")
plt.ylabel("Final Grade")
plt.xlabel("StudyTime(hours)")
plt.legend(title="Gender")
plt.show()

#3. Create a bar chart comparing the average scores of male and female students.
plt.figure(figsize=(8,5))
avg_grade.plot(kind='bar',color=['blue','pink'])
plt.title("Average Final Grade By Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.xticks(rotation=0)
plt.show
```
